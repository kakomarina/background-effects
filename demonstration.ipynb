```python
from triangle_threshold import median_filter, triangle_threshold, histogram
from change_background import change_background
from background_blur import background_blur
import imageio
import matplotlib.pyplot as plt
from skimage import color
```

```python
girl_original = imageio.imread("girl1.jpg")
girl_grayscale = imageio.imread("girl1.jpg", as_gray=True)


plt.subplot(1, 2, 1)
plt.imshow(girl_original)
plt.title('girl')
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(girl_grayscale, cmap='gray')
plt.title('grayscale image')
plt.axis('off')
```

```python
background = imageio.imread("bg_mata.jpg")

plt.imshow(background)
```

```python
image_filtered = median_filter(girl_grayscale)
```

```python
hist = histogram(image_filtered)

plt.bar(range(255), hist)
```

```python
threshold = triangle_threshold(image_filtered)
bool_image = image_filtered > threshold
```

```python
plt.imshow(bool_image, cmap='gray')
```

```python
final_image = change_background(girl_original, background, bool_image)
```

```python
plt.imshow(final_image)
```
